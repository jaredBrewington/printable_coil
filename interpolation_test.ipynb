```python
from scipy import interpolate
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
import numpy as np
```

```python
s = np.arange(0,10)
x = np.random.randint(0,5,size=len(s))
y = np.random.randint(0,5,size=len(s))
z = np.arange(0,10,(10/len(s)))
```

```python
spl_x = interpolate.InterpolatedUnivariateSpline(s,x)
spl_y = interpolate.InterpolatedUnivariateSpline(s,y)
spl_z = interpolate.InterpolatedUnivariateSpline(s,z)
```

```python
ss = np.arange(0,s.max(), .01)
xs = spl_x(ss)
ys = spl_y(ss)
zs = spl_z(ss)

plt.figure(figsize=(15,10))
plt.plot(s,x,'o')
plt.plot(ss,xs)

plt.figure(figsize=(15,10))
plt.plot(s,y,'o')
plt.plot(ss,ys)

plt.figure(figsize=(15,10))
plt.plot(s,z,'o')
plt.plot(ss,zs)

plt.show()
```

```python
%matplotlib notebook
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot(x,y,z, 'o')
ax.plot(xs, ys, zs)
```
