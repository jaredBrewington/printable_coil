```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from wire import Wire
import numpy as np
```

```python
wire = Wire("prototype_double_cos_coil/coil1/wires/wire10.txt")
```

```python
wire.points.shape
```

```python
dx = []
dy = []
dz = []
for i in np.arange(wire.points.shape[1])[::20]:
    d_vec = wire.interp.derivative_direction(i)
    dx.append(d_vec[0])
    dy.append(d_vec[1])
    dz.append(d_vec[2])
```

```python
%matplotlib notebook
fig = plt.figure(figsize=(5,5))
ax = plt.axes(projection="3d")
ax.plot(wire.points[0], wire.points[1], wire.points[2], 'o-', markersize=1, c='k')
ax.quiver(wire.points[0][::20], wire.points[1][::20], wire.points[2][::20], dx, dy, dz, length=0.01)
range = max(wire.points[0]) - min(wire.points[0])
ax.set_ylim(-range/2, range/2)
ax.set_zlim(-range/2, range/2)
ax.set_xlabel("x")
ax.set_ylabel("y")
```
